```python
from pyspark.sql import SparkSession
spark= SparkSession.builder.appName("Missing").getOrCreate()
```

```python
df=spark.read.csv(r"C:\Users\faisal\Downloads\breast cancer.csv",header=True,inferSchema=True)
df.show()
```

```python
df.printSchema()
```

```python
df.columns
```

```python
from pyspark.ml.feature import VectorAssembler
feature_assembler=VectorAssembler(inputCols=['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave points_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst'],outputCol='Independent Feature')
```

```python
output=feature_assembler.transform(df)
output.show()
```

```python
output.columns
```

```python
finalised_data=output.select('Independent Feature','diagnosis')
finalised_data.show()
```

```python
from pyspark.ml.feature import StringIndexer
indexer=StringIndexer(inputCol="diagnosis",outputCol="Index")
#fits a model to the input dataset with optional parameters
df1=indexer.fit(finalised_data).transform(finalised_data)
df1=df1.select("Independent Feature","Index")
df1.show()
```

```python
from pyspark.ml.classification import LogisticRegression
#train test split
train_data,test_data=df1.randomSplit([0.75,0.25])
#train the logistic regresion model
regressor=LogisticRegression(featuresCol='Independent Feature',labelCol='Index')
regressor=regressor.fit(df1)
```

```python
test_data.show()
```

```python
#prediction
pred_results=regressor.evaluate(test_data)
```

```python
pred_results.predictions.show()
pd=pred_results.predictions.select('Index','prediction')
```

```python
pred_labels=pred_results.predictions.select('Index','prediction')
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator,MulticlassClassificationEvaluator
evaluator=BinaryClassificationEvaluator(rawPredictionCol='prediction',labelCol='Index')
evaluator=MulticlassClassificationEvaluator(predictionCol='prediction',labelCol='Index',metricName='accuracy')
```
